### Read Data

```python
df = pd.read_csv('./Dataset/test_data1.csv', header = 0, sep =',')
df.name = 'test_data'

df.head(1)
```

### Basic operations

```python
df[df == 'na'] = pd.np.nan
```

```python
#Data info
# df.info()
df.shape
```

```python
# Replace whitespace in Columns and convert to Lower case
df.columns = df.columns.str.lower().str.replace(' ', '_')
# df.info()
```

```python
#Change dtypes
df["payment_default"] = df["payment_default"].astype('category')
df[df.select_dtypes(include='int64').columns.values] = df[df.select_dtypes(include='int64').
                                                              columns.values].astype(float)
```

```python
#define Basic features
unique_feature = 'id'
target_var = 'payment_default'
```

```python
#Extract feature names for categorical and numeric data types
categorical_features = df.select_dtypes(include='object').columns.values
numerical_features = df.select_dtypes(include='float64').columns.values
```

```python
# Drop Rows with all missing values
df.dropna(axis=0, how='all', inplace = True)

#Drop Rows with empty Identifier
df = df.loc[(df[unique_feature]!='') & (df[unique_feature]!= np.nan), ]

#Drop Rows with empty target_var
df = df.loc[(df[target_var]!='') & (df[target_var]!= np.nan), ]
```

### Missing Values
- http://localhost:8888/notebooks/ML%20Concepts/Data%20Transformation/Data%20Exploration/Python%20Code/Missing%20Values.ipynb

```python
# print(df.isna().sum())
# print(df_most_frequent_imputation.isna().sum())
# set(df_most_frequent_imputation['married_status'])
# set(df['married_status'])
```

```python
#Using most_frequent
df_most_frequent_imputation = df.copy()

imputer = Imputer(missing_values = 'NaN', strategy ='most_frequent', axis = 0)
for feature in numerical_features:
    df_most_frequent_imputation[feature]= imputer.fit_transform(df_most_frequent_imputation[[feature]]).ravel().round()

#Using mode
for feature in categorical_features:
    df_most_frequent_imputation[feature].fillna(df_most_frequent_imputation[feature].mode()[0], inplace=True)

# print(df_most_frequent_imputation.isna().sum())
df_most_frequent_imputation.to_csv('df_most_frequent_imputation.csv', index= False)
```

#### Compare two data series using this Matplotlib

```python
#Data to plot
n_groups = 4
A = [90, 55, 40, 65]
B = [85, 62, 54, 20]

#Create plots
fig, ax = plt.subplots()
index = np.arange(n_groups)
barwidth = 0.35
opacity = 0.6

rects1 = plt.bar(index, A, barwidth, 
                 alpha = opacity,
                 color ='b',
                 label = 'A'
                )

rects1 = plt.bar(index + barwidth, B, barwidth, 
                 alpha = opacity,
                 color ='g',
                 label = 'B'
                )

plt.xlabel('Persons')
plt.ylabel('Scores')
plt.title("Persons with Scores")
plt.xticks(index+barwidth, ['A1', 'A2', 'A3', 'A4'])

plt.legend()
plt.tight_layout()

plt.show()
```

##### Sample code - Stacked Bar Graph

```python
#Prepare data
N = 5
men_means = [20, 35, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

menstd = [2, 3, 4, 1, 2]
womenstd = [3, 5, 2, 3, 3]

ind = np.arange(N)    # the x locations for the groups
barwidth = 0.35       # the width of the bars: can also be len(x) sequence
opacity = 0.7

p1 = plt.bar(ind, men_means, barwidth, color = 'b', alpha = opacity, yerr = menstd)
p2 = plt.bar(ind, women_means, barwidth, bottom = men_means, color = 'g', alpha = opacity, yerr = womenstd)

plt.ylabel('Scores')
plt.title('Scores by group and gender')

plt.xticks(ind, ['G1', 'G2', 'G3', 'G4', 'G5'])
plt.yticks(np.arange(0, 81, 10))

plt.legend((p1[0], p2[0]), ['Men', 'Women'])
plt.show()
```

##### Sub plots

```python
# Simple data to display in various forms
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

plt.close('all')

# # Just a figure and one subplot
f, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Simple plot')

plt.show()
```

```python
# Two subplots, the axes array is 1-d
f, axarr = plt.subplots(2, sharex=True)
# f, axarr = plt.subplots(2, sharex='col')

axarr[0].plot(x, y)
axarr[0].set_title('Sharing X axis')
axarr[1].scatter(x, y)

plt.show()
```

```python
# Two subplots, unpack the axes array immediately
# f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
f, axarr = plt.subplots(1, 2, sharey='row')
axarr[0].plot(x, y)
axarr[0].set_title('Sharing Y axis')
axarr[1].scatter(x, y)

plt.show()
```

```python
# Three subplots sharing x axes
f, axarr = plt.subplots(3, sharex=True)
axarr[0].plot(x, y)
axarr[0].set_title('Sharing both axes')
axarr[1].scatter(x, y)
axarr[2].scatter(x, 2 * y ** 2 - 1, color='r')

# Fine-tune figure; make subplots close to each other and hide x ticks for
# all but bottom plot.

f.subplots_adjust(hspace=0)
plt.setp([a.get_xticklabels() for a in f.axes[:-1]], visible=False)

plt.show()
```

```python
# row and column sharing
f, axarr = plt.subplots(2, 2, sharex='col', sharey='row')
axarr[0,0].plot(x, y)
axarr[0,0].set_title('Sharing x per column, y per row')

axarr[1,0].scatter(x, y)
axarr[0,1].plot(x, 2 * y ** 2 - 1, color='r')
axarr[1,1].scatter(x, 2 * y ** 2 - 1, color='r')

plt.show()
```

```python
from scipy.stats import trim_mean, kurtosis, variation 
from scipy.stats.mstats import mode, gmean, hmean
import pandas as pd
```

```python
target_varible = 'payment_default'
test_variable = 'age'
```

```python
def summarize_df1(df, target_varible, test_variable):
        from scipy.stats import trim_mean, kurtosis, variation 
        from scipy.stats.mstats import mode, gmean, hmean
        import pandas as pd

        train_grouped_data = df.groupby([target_varible])
        descr = train_grouped_data[test_variable].describe().reset_index()
        
        descr['IQR'] = descr['75%'] - descr['25%']
        
        total = train_grouped_data[test_variable].sum().reset_index()
        descr['total'] = pd.Series(total[test_variable].values, index=descr.index)
        
        unique_count = train_grouped_data[test_variable].nunique().reset_index()
        descr['unique_count'] = pd.Series(unique_count[test_variable].values, index=descr.index)
        
        median = train_grouped_data[test_variable].median().reset_index()
        descr['median'] = pd.Series(median[test_variable].values, index=descr.index)
        
        mode = train_grouped_data[test_variable].apply(mode, axis=None).reset_index()
        descr['mode'] = pd.Series(mode[test_variable].values, index=descr.index)
        
        var = train_grouped_data[test_variable].var().round(1).reset_index()
        descr['var'] = pd.Series(var[test_variable].values, index=descr.index)
        
        cv = train_grouped_data[test_variable].apply(variation, axis=None).reset_index()
        descr['cv'] = pd.Series(cv[test_variable].values, index=descr.index)

        trimmed_mean = train_grouped_data[test_variable].apply(trim_mean, .1).reset_index()
        descr['trimmed_mean'] = pd.Series(trimmed_mean[test_variable].values, index=descr.index)
        
        descr['harmonic_mean'] = ''
        descr['geo_mean'] = ''
        
        nonzero = True
        for key, item in train_grouped_data:
            if(sum(item[test_variable] <=0 )):
                nonzero = False

        if(nonzero):
            mean = train_grouped_data[test_variable].apply(hmean, axis=None).reset_index()
            descr['harmonic_mean'] = pd.Series(mean[test_variable].values, index=descr.index)
            
            mean = train_grouped_data[test_variable].apply(gmean, axis=None).reset_index()
            descr['geo_mean'] = pd.Series(mean[test_variable].values, index=descr.index)

        else:
            descr.loc[:,'harmonic_mean'] = 'Non Positive'
            descr.loc[:,'geo_mean'] = 'Non Positive'
        
        skewness = train_grouped_data[test_variable].skew().reset_index()
        descr['skewness'] = pd.Series(skewness[test_variable].values, index=descr.index)

        kurtosis = train_grouped_data[test_variable].apply(pd.DataFrame.kurt).reset_index()
        descr['kurtosis'] = pd.Series(kurtosis[test_variable].values, index=descr.index)

        return descr
    
```
